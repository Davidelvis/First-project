# VISUALIZATION OF COVID 19 DATA FROM INDIA USING A MAP.#

You can a world map plus folium library to visualize and analyse data. And we are going to explore step by step until we plot a map.

```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
#style.use("ggplot")
%matplotlib inline

import plotly
import plotly.express as px
import plotly.graph_objects as go
#to set the size of the fisures that will be created.
plt.rcParams["figure.figsize"]=10,8
import cufflinks as cf
import plotly.offline as pyo
from plotly.offline import init_notebook_mode,plot,iplot
import folium
```

```python
pyo.init_notebook_mode(connected=True)
cf.go_offline()
```

```python
# Reading data.
dfindia=pd.read_excel("Covidcasesindia.xlsx")
# Dropping the S.No. column
dfindia.drop(["S. No."],axis=1, inplace=True)
#Creating a new column by adding two columns.
dfindia["Total cases"]=dfindia["Total Confirmed cases (Indian National)"]+dfindia["Total Confirmed cases ( Foreign National )"]
#Adding a new column
dfindia["Active cases"]=dfindia["Total cases"]-(dfindia["Death"]+dfindia["Cured"])
india_cord=pd.read_excel("Indian Coordinates.xlsx.")
```

```python
#Reading data
india_cord
```

```python
#We will need to merge the india covid-19 data with the data of states coordinates
df_full= pd.merge(india_cord,dfindia,on="Name of State / UT")
```

```python
#reading the merged dataset
df_full
```

```python
map=folium.Map(location=[20,70],zoom_start=4,tiles="stamen terrain")

for lat,long,value, name in zip(df_full["Latitude"],df_full["Longitude"],df_full["Total cases"],df_full["Name of State / UT"]):
    folium.CircleMarker([lat,long],radius=value*0.3,tooltip="click Me",popup=("<strong>state</strong>: "+str(name).capitalize()+'<br>''<strong>Total cases</strong>: '+str(value)+ '<br>'),color='red',fill_color='red',fill_opacity=0.2).add_to(map)
```
